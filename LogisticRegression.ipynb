**Data Cleaning**

```python
df.isnull().sum().to_frame().style.set_properties(
    **{
    'background-color' : 'purple',
    'color':'white',
    'border-color':'white',
    'border-style':'solid'

}
)
```

```python
col_list = df.columns
```

```python
for x in col_list:
  if df[x].dtypes =='object':
    df[x] = df[x].fillna(df[x].mode()[0])
  else:
    df[x] = df[x].fillna(df[x].mean())
```

```python
df.isnull().sum().to_frame().style.set_properties(
    **{
    'background-color' : 'purple',
    'color':'white',
    'border-color':'white',
    'border-style':'solid'

}
)
```

```python
oultier_col = ['age', 'duration','campaign','cons.conf.idx']
```

```python
for x in oultier_col:
  Q1 = df[x].quantile(0.25)
  Q3 = df[x].quantile(0.75)
  IQR = Q3-Q1

  LL =    Q1-1.5*IQR
  UU =    Q3+1.5*IQR

  df = df[(df[x] >= LL) & (df[x] <=UU)]

```
