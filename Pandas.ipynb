**Merge / Join**

```python
DF1 = pd.DataFrame({"ID": [1, 2, 3], "Name": ["A", "B", "C"]})
DF2 = pd.DataFrame({"ID": [1, 2, 4], "Score": [90, 80, 70]})
```

```python
pd.merge(DF1,DF2,on="ID")
#DF1.merge(DF2, on="ID"
```

```python
pd.merge(DF1,DF2,on="ID",how="left")
#DF1.merge(DF2, on="ID", how="left")
```

```python
pd.concat([DF1,DF2])
```

```python
DF3 = pd.DataFrame({"A": [1, 2, 3]}, index=["x", "y", "z"])
DF4 = pd.DataFrame({"B": [4, 5, 6]}, index=["x", "y", "z"])
```

1. Join on index (default)

```python
DF3.join(DF4)
```

2. Join with different indexes

```python
DF5 = pd.DataFrame({"C": [7, 8]}, index=["x", "y"])
DF3.join(DF5, how="left")
```

3. Join on a column instead of index

```python
DF6 = pd.DataFrame({"key": ["x", "y", "z"], "D": [10, 11, 12]})
DF3.reset_index().join(DF6.set_index("key"), on="index")
```

- If you see a TypeError, it’s usually because:

  - You passed something that isn’t a DataFrame/Series to join().

  - You didn’t align indexes properly.

  - You tried to join on a column without resetting/setting the index.

**Pivot Tables**

`pivot()` - Reshape data by turning unique values from one column into new columns.

```python
DF = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT"],
    "Year": [2024, 2025, 2024, 2025],
    "Salary": [50000, 55000, 60000, 65000]
})

DF
```

```python
DF.pivot(index="Department", columns="Year", values="Salary")
```

`pivot_table()` - Similar to `pivot()`, but handles duplicates by applying an aggregation function.

```python
DF.pivot_table(index="Department", columns="Year", values="Salary", aggfunc="mean")
```

```python
DF1 = pd.DataFrame({
    "Region": ["North", "South", "North", "South", "East"],
    "Year": [2024, 2024, 2025, 2025, 2024],
    "Sales": [100, 200, 150, 250, 300]
})

DF1
```

```python
pd.pivot_table(DF1, values="Sales", index="Region")
```

```python
#Change aggregation function:
pd.pivot_table(DF1, values="Sales", index="Region", aggfunc="sum")
```

```python
#Multiple aggregations:
pd.pivot_table(DF1, values="Sales", index="Region", aggfunc=["mean", "sum", "max"])
```

```python
#Add columns for grouping:
pd.pivot_table(DF1, values="Sales", index="Region", columns="Year", aggfunc="sum")
```
